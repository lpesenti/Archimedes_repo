# ET Noise Budget Curves

These are plots of the noise budget curves and examples on how to work with them.

To plot these curves yourself, you will need:

* an up-to-date python installation (e.g. Anaconda)
* pygwinc (`pip install gwinc`)

```python
import matplotlib.pyplot as plt
import gwinc 
import numpy as np
%matplotlib inline 
```

## ET High-Frequency

```python
budgetHF = gwinc.load_budget('ETHF')
tracesHF = budgetHF.run()
fig = gwinc.plot_budget(tracesHF)
plt.ylabel('Strain [[$1 / \sqrt{\mathrm{Hz}}$]]')
plt.ylim([5e-26, 1e-22])
```

## ET Low-Frequency (two material coating)

```python
budgetLF = gwinc.load_budget('ETLF')
#budgetLF.ifo.Seismic.Site='SosEnattos'
tracesLF = budgetLF.run()
fig = gwinc.plot_budget(tracesLF)
plt.ylabel('Strain [[$1 / \sqrt{\mathrm{Hz}}$]]')
plt.legend(loc='upper right')
plt.ylim([5e-26, 1e-21])
```

## ET Low-Frequency (four material coating)

```python
budgetLFmmc = gwinc.load_budget('ETLF_mmc')
tracesLFmmc = budgetLFmmc.run()
fig = gwinc.plot_budget(tracesLFmmc)
plt.ylabel('Strain [[$1 / \sqrt{\mathrm{Hz}}$]]')
plt.legend(loc='upper right')
plt.ylim([5e-26, 1e-21])
```

## ET Combined

```python
budgetET = gwinc.load_budget('ET')
tracesET = budgetET.run()
fig = gwinc.plot_budget(tracesET,)
plt.ylabel('Strain [[$1 / \sqrt{\mathrm{Hz}}$]]')
plt.title('Curent model of ET NB ');
```
