```python
import inspiral_range
import gwinc
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data=np.loadtxt("ETSensitivity_Scenarios.txt")
fVh=Data[:,0]
NVh=Data[:,1]
fVl=Data[:,2]
NVl=Data[:,3]
fA=Data[:,4]
NA=Data[:,5]
fET=Data[:,12]
NET=Data[:,13]
```

```python
M=np.linspace(0.5,500,600)
```

```python
redETD=np.empty([600])
redAplus=np.empty([600])
redAVirgoH=np.empty([600])
redAVirgoL=np.empty([600])
```

```python
for i in range(len(M)):
    rangeAVirgoH = inspiral_range.horizon_redshift(
    fVh,NVh**2 ,
    m1=M[i], m2=M[i])
    redAVirgoH[i]=rangeAVirgoH


for i in range(len(M)):
    rangeAVirgoL = inspiral_range.horizon_redshift(
    fVl, NVl**2,
    m1=M[i], m2=M[i])
    redAVirgoL[i]=rangeAVirgoL



for i in range(len(M)):
    rangeAp = inspiral_range.horizon_redshift(
    fA, NA**2,
    m1=M[i], m2=M[i])
    redAplus[i]=rangeAp


for i in range(len(M)):
    rangeETD = inspiral_range.horizon_redshift(
    fET, NET**2,
    m1=M[i], m2=M[i])
    redETD[i]=rangeETD
```

```python
fig = plt.figure(figsize=(7,4.3))

plt.loglog(2*M,redAVirgoH,lw=2)
plt.loglog(2*M,redAVirgoL,lw=2)
plt.loglog(2*M,redAplus,lw=2)
plt.loglog(2*M,redETD,lw=2)

plt.grid(which='both',ls='-',linewidth=0.5,alpha=0.5)
plt.legend(['Advanced Virgo+ (high)','Advanced Virgo+ (low)','Advanced LIGO+','ET design'],ncol=2,loc='lower right')
plt.xlabel('Total Mass [$M_\odot$]')
plt.ylabel('Redshift')
plt.xlim(1,1000)
```
